```python
import cnn,time
import pandas as pd
```

```python
def single_test(country,df):
    a = time.time()
    pred,MAPE,figure = cnn.predict(df,country)
    b = time.time()
    print(f"time taken for model to run is {round(b-a,2)} and error of model is {MAPE}")
    print(pred)
    figure.show()
```

```python
single_test("India","recovered")
```

```python
def country_test(country):
    df = pd.DataFrame()
    errors = []
    times = []
    pred_lists = []
    df["Type"] = ["confirmed","recovered","deaths"]
    for i in df.Type:
        b = time.time()
        pred,MAPE,figure = cnn.predict(i,country)
        errors.append(MAPE)
        a = time.time()
        pred_lists.append(pred)
        times.append(round(a-b,2))
    df["times"] = times
    df["errors"] = errors
    df["pred"] = pred_lists
    return df

```

```python
country_test("Tanzania")
```
